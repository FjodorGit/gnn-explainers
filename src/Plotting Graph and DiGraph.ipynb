```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.DiGraph()

G.add_nodes_from([1,2])
G.add_edge(1,2)
G.add_edge(2,1)

nx.draw_networkx(G)

plt.show()
```

```python
G = nx.Graph()

G.add_nodes_from([1,2])
G.add_edge(1,2)
G.add_edge(2,1)

nx.draw_networkx(G)

plt.show()
```
